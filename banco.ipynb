```python
import pandas as pd
import chardet

# Specify relative path to the file
file_path = "../IAPS/banco_completo.csv"

# Detect the encoding of the file
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())
    file_encoding = result['encoding']

# Load data into a pandas DataFrame with the correct encoding and delimiter specified
daf = pd.read_csv(file_path, encoding=file_encoding, delimiter=';')


```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming you already have 'santa_casa_rows' and 'not_santa_casa_rows' DataFrames

# Calculate the contingency table
contingency_table = pd.crosstab(index=daf['obito'], columns=daf['is_santa_casa'])

# Perform the chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Statistic:", chi2)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming you already have 'daf' DataFrame loaded from the CSV file using the correct encoding and delimiter

# Get the unique values in the 'estabel_saude' column
unique_estabel_saude = daf['estabel_saude'].unique()

# Create a dictionary to store the results
p_values_dict = {}

# Iterate through the unique values and perform chi-square test for each one
for estabel_value in unique_estabel_saude:
    # Create a new column to differentiate the rows for the current 'estabel_value'
    daf[f'is_{estabel_value.replace(" ", "_").lower()}'] = daf['estabel_saude'].apply(lambda x: x == estabel_value)

    # Separate rows based on the current 'estabel_value' and 'obito'
    contingency_table = pd.crosstab(index=daf['obito'], columns=daf[f'is_{estabel_value.replace(" ", "_").lower()}'])

    # Perform the chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)

    # Store the p-value in the dictionary
    p_values_dict[estabel_value] = p_value

# Find the unique values with p-value less than 0.05
significant_values = [value for value, p_value in p_values_dict.items() if p_value < 0.05]

# Print the significant values
print("Unique items in 'estabel_saude' with p-value < 0.05:")
print(significant_values)
```

```python
import numpy as np

def cramers_v(confusion_matrix):
    """ calculate Cramers V statistic for categorial-categorial association. """
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))

# ... (your previous code up to for loop)

for estabel_value in unique_estabel_saude:
    # ... (your previous code inside the for loop)

    # Perform the chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)

    # Calculate Cramer's V
    cramers_v_value = cramers_v(contingency_table)

    # Store the p-value and Cramer's V in the dictionary
    p_values_dict[estabel_value] = {"p_value": p_value, "cramers_v": cramers_v_value}

# ... (your previous code after the for loop)

# Print the significant values and their corresponding Cramer's V
print("Unique items in 'estabel_saude' with p-value < 0.05 and their Cramer's V:")
for value, stats in p_values_dict.items():
    if stats["p_value"] < 0.05:
        print(f"{value}: Cramer's V = {stats['cramers_v']}")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Assuming you already have 'df' DataFrame loaded from the CSV file using the correct encoding and delimiter

# Get the unique values in the 'estabel_saude' column
unique_estabel_saude = df['estabel_saude'].unique()

# Create a dictionary to store the results
p_values_dict = {}
cramers_v_dict = {}

# Define a function for Cramer's V
def cramers_v(chi2, n, shape):
    return np.sqrt(chi2 / (n*(min(shape)-1)))

# Iterate through the unique values and perform chi-square test for each one
for estabel_value in unique_estabel_saude:
    # Create a new DataFrame for the current 'estabel_value'
    df_temp = df[df['estabel_saude'] == estabel_value]

    # Separate rows based on the current 'estabel_value' and 'obito'
    contingency_table = pd.crosstab(index=df_temp['obito'], columns=df_temp['estabel_saude'])

    # Perform the chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)

    # Store the p-value in the dictionary
    p_values_dict[estabel_value] = p_value

    # Calculate and store Cramer's V in the dictionary
    if p_value < 0.05:
        cramers_v_dict[estabel_value] = cramers_v(chi2, len(df_temp), contingency_table.shape)

# Print the significant values and their CramÃ©r's V
for key in cramers_v_dict:
    print(f"{key}: Cramer's V = {cramers_v_dict[key]}")
```
