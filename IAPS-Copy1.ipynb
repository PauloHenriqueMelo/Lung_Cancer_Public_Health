```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame called "df"
# Replace "obito" and "tempo_internacao" with the actual column names in your DataFrame

# Split the dataset into two groups based on obito
group_1 = df[df["obito"] == 1]["tempo_internacao"]
group_0 = df[df["obito"] == 0]["tempo_internacao"]

# Define the number of bootstrap iterations
n_iterations = 1000

# Initialize arrays to store the means and confidence intervals
means_1 = np.zeros(n_iterations)
means_0 = np.zeros(n_iterations)
ci_1 = np.zeros((n_iterations, 2))
ci_0 = np.zeros((n_iterations, 2))

# Perform bootstrap sampling and calculate means and confidence intervals for each group
for i in range(n_iterations):
    # Generate bootstrap samples for each group
    bootstrap_sample_1 = np.random.choice(group_1, size=len(group_1), replace=True)
    bootstrap_sample_0 = np.random.choice(group_0, size=len(group_0), replace=True)
    
    # Calculate means for each bootstrap sample
    means_1[i] = np.mean(bootstrap_sample_1)
    means_0[i] = np.mean(bootstrap_sample_0)
    
    # Calculate confidence intervals using the percentile method for each group
    ci_1[i] = np.percentile(bootstrap_sample_1, [2.5, 97.5])
    ci_0[i] = np.percentile(bootstrap_sample_0, [2.5, 97.5])

# Plot the distributions of means and confidence intervals for each group
plt.hist(means_1, bins=30, alpha=0.5, label="Obito 1")
plt.hist(means_0, bins=30, alpha=0.5, label="Obito 0")
plt.axvline(np.mean(means_1), color='b', linestyle='-', label='Mean (Obito 1)')
plt.axvline(np.mean(means_0), color='g', linestyle='-', label='Mean (Obito 0)')
plt.axvline(np.mean(ci_1[:, 0]), color='r', linestyle='--', label='CI Lower (Obito 1)')
plt.axvline(np.mean(ci_1[:, 1]), color='r', linestyle='--', label='CI Upper (Obito 1)')
plt.axvline(np.mean(ci_0[:, 0]), color='m', linestyle='--', label='CI Lower (Obito 0)')
plt.axvline(np.mean(ci_0[:, 1]), color='m', linestyle='--', label='CI Upper (Obito 0)')
plt.xlabel("Mean of Tempo de Internação")
plt.ylabel("Frequency")
plt.title("Bootstrap Distributions of Means")
plt.legend()
plt.show()

# Print the mean and confidence intervals for each group
print("Mean (Obito 1):", np.mean(means_1))
print("95% Confidence Interval (Obito 1):", np.mean(ci_1, axis=0))
print("Mean (Obito 0):", np.mean(means_0))
print("95% Confidence Interval (Obito 0):", np.mean(ci_0, axis=0))

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have group_1 and group_2 as numpy arrays
# For demonstration purposes, I'll use random data
# Split the dataset into two groups based on obito
group_1 = df[df["obito"] == 1]["tempo_internacao"]
group_2 = df[df["obito"] == 0]["tempo_internacao"]
def bootstrap(data, n_bootstrap=1000):
    n = len(data)
    means = np.empty(n_bootstrap)
    for i in range(n_bootstrap):
        sample = np.random.choice(data, n)  # this performs the resampling step
        means[i] = np.mean(sample)  # calculates the mean of the sample
    ci_lower = np.percentile(means, 2.5)  # lower percentile for 95% CI
    ci_upper = np.percentile(means, 97.5)  # upper percentile for 95% CI
    return means, ci_lower, ci_upper

# bootstrap group_1 and group_2
bootstrap_means_group1, ci_l_group1, ci_u_group1 = bootstrap(group_1)
bootstrap_means_group2, ci_l_group2, ci_u_group2 = bootstrap(group_2)

# Print the mean and 95% CI for each group
print(f'Group 1: Mean = {np.mean(group_1):.2f}, 95% CI = [{ci_l_group1:.2f}, {ci_u_group1:.2f}]')
print(f'Group 2: Mean = {np.mean(group_2):.2f}, 95% CI = [{ci_l_group2:.2f}, {ci_u_group2:.2f}]')

# Plot the bootstrap distributions of the means
plt.figure(figsize=(10,6))

sns.histplot(bootstrap_means_group1, color='blue', kde=True, label='Group 1')
plt.axvline(x=ci_l_group1, color='blue', linestyle='--')
plt.axvline(x=ci_u_group1, color='blue', linestyle='--')

sns.histplot(bootstrap_means_group2, color='green', kde=True, label='Group 2')
plt.axvline(x=ci_l_group2, color='green', linestyle='--')
plt.axvline(x=ci_u_group2, color='green', linestyle='--')

plt.xlabel('Mean')
plt.ylabel('Frequency')
plt.legend()
plt.title('Bootstrap Distributions of Mean with 95% Confidence Intervals')
plt.show()
```
